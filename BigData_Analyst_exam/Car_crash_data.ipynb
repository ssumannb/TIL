```python
import seaborn as sns
import pandas as pd

sns.get_dataset_names()
df = sns.load_dataset('car_crashes')
df.head()
```

Q1. alcohol(음주수치)이 상위 10번째 값으로 상위 1~10위의 범위의 값을 변경한 후  
speeding(속도)가 7 이상의 alcohol 데이터들의 평균 산출


```python
df_sorted = df.sort_values(by='alcohol', ascending=False)
df_sorted.head(10)
```

```python
df_sorted['alcohol'].iloc[:10] = df_sorted['alcohol'].iloc[9]
df_speed = df_sorted[df_sorted['speeding']>=7.0]
print(df_speed['alcohol'].mean())
```
